```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
from sympy import init_printing
from IPython.display import display, Math

init_printing()
x = sp.Symbol('x')
f = sp.exp(-x**2)
display(Math(r"f(x) = e^{-x^2}"))
display(Math(r"\int_{-\infty}^{\infty} e^{-x^2} \, dx"))
res = sp.integrate(f, (x, -sp.oo, sp.oo))

display(Math(r"Result = " + sp.latex(res)))

approx = res.evalf()
display(Math(r"\text{Approximate value} \approx " + str(approx)))

x_vals = np.linspace(-5, 5, 400)
y_vals = np.exp(-x_vals**2)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label=r"$e^{-x^2}$", color="blue")
plt.fill_between(x_vals, y_vals, alpha=0.3, color="blue")
plt.title("Integral of the Gaussian Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
